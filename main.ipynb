```python
import pandas as pd;
import numpy as np;
import matplotlib.pyplot as plt;
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression;

#load csv file 
dt=pd.read_csv('Salary.csv');
```

```python
# using numpy function iloc[row,coloumn] 

real_x=dt.iloc[:,0].values;
real_y=dt.iloc[:,1].values;

#convert in 2d array

real_x=real_x.reshape(-1,1);
real_y=real_y.reshape(-1,1);
```

```python
#using sklearn funtion train_test_split suncyion
training_x,testing_x,training_y,testing_y=train_test_split(real_x,real_y,test_size=0.3,random_state=0)
```

```python
# using LinearRegression for prediction 
lin=LinearRegression();
lin.fit(training_x,training_y);
pre=lin.predict(testing_x);
```

```python
# show in graph  using matplotlib.pyplot  

plt.scatter(training_x,training_y,color='red')
plt.plot(training_x,lin.predict(training_x),color='green')
plt.title('salary & exp Training object');
plt.xlabel('exp');
plt.ylabel('salary');
plt.show()
```

```python
plt.scatter(testing_x,testing_y,color='red')
plt.plot(training_x,lin.predict(training_x),color='green')
plt.title('salary & exp Training object');
plt.xlabel('exp');
plt.ylabel('salary');
plt.show()
```
